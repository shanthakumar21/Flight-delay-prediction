```
import pandas as pd
import os
import numpy as np  
```

```
f1=pd.read_csv("gdrive/My Drive/finalflight.csv")
w=pd.read_csv("gdrive/My Drive/weatherdata.csv")
```

```
f1["ArrFlightDate"]=f1["FlightDate"]
f1.ArrFlightDate=pd.to_datetime(f1.ArrFlightDate)
f1.FlightDate=pd.to_datetime(f1.FlightDate)
f1["ArrAbsTime"]=f1["CRSArrTime"]
f1
```

```
for index, row in f1.iterrows():
    lastdigits=row["ArrAbsTime"]%100
    absolutetime=row["ArrAbsTime"]-lastdigits
    f1.at[index,"ArrAbsTime"]=absolutetime
    if(row["CRSDepTime"]>row["CRSArrTime"]):
        f1.at[index,"ArrFlightDate"]= row["FlightDate"]+pd.DateOffset(days=1)
        
```

```
f1
```

```
f1=f1.dropna()
```

```
f1['ArrAbsTime'] = f1['ArrAbsTime'].astype(np.int64)

f1["DepAbsTime"]=f1["CRSDepTime"]
for index, row in f1.iterrows():
    lastdigits=row["DepAbsTime"]%100
    absolutetime=row["DepAbsTime"]-lastdigits
    f1.at[index,"DepAbsTime"]=absolutetime
f1 
```

```
f1['DepAbsTime'] = f1['DepAbsTime'].astype(np.int64)
w.date=pd.to_datetime(w.date)
```

```
for index,row in f1.iterrows():
    if(row["CRSArrTime"]==2400):
        f1.at[index,"CRSArrTime"]=0
        f1.at[index,"ArrAbsTime"]=0
```

```
count = 0
for index,row in f1.iterrows():
    if(row["CRSArrTime"]==0):
        count = count + 1
print(count) 
```
