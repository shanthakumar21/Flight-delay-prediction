```
m2 = pd.read_csv("/content/mod3manipulation.csv")
```

```
m2
```

```
X=m2.loc[m2['Arrdel15'] == 1.0]      #selecting only delayed flights 
X

```

```
Y = X['ArrDelayMinutes']
Y
```

```
X.shape
```

```
Y.shape
```

```
X=X.drop(['ArrDelayMinutes','Unnamed: 0'],axis = 1)
```

```
X.transpose()
```

```
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20) 
```

```
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```
##linear regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```
Y_pred_lin_test=reg.predict(X_test)
```

```
print("mean_squared_error   " , np.sqrt(mean_squared_error(Y_test, Y_pred_lin_test)))
print("mean_absolute_error  " , mean_absolute_error(Y_test, Y_pred_lin_test))
print("R2score    " , r2_score(Y_test,Y_pred_lin_test))
```

```
## EXTRATREESREGRESSOR Tree
from sklearn.ensemble import ExtraTreesRegressor
etr=ExtraTreesRegressor()
etr.fit(X_train,Y_train)
```

```
Y_pred_etr_test = etr.predict(X_test)
```

```
print("mean_squared_error   " , np.sqrt(mean_squared_error(Y_test, Y_pred_etr_test)))
print("mean_absolute_error  " , mean_absolute_error(Y_test, Y_pred_etr_test))
print("R2score    " , r2_score(Y_test,Y_pred_etr_test))
```

```
#XGB Regressor
import xgboost as xgb
xgbmodel = xgb.XGBRegressor()
xgbmodel.fit(X_train,Y_train)
```

```
Y_pred_xgbr = xgbmodel.predict(X_test)
```

```
print("mean_squared_error   " , np.sqrt(mean_squared_error(Y_test, Y_pred_xgbr)))
print("mean_absolute_error  " , mean_absolute_error(Y_test, Y_pred_xgbr))
print("R2score    " , r2_score(Y_test,Y_pred_xgbr))
```
